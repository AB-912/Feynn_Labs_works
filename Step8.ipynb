```python
import numpy as np
import matplotlib.pyplot as plt
```

**The upper half of the decision matrix as matrix x and the corresponding weights from the first column as vector wx**

```python
x = np.array([
    [5, 10, 1, 5, 10, 3, 1, 10],
    [2, 1, 2, 6, 9, 4, 2, 10],
    [10, 6, 4, 4, 8, 2, 1, 9],
    [8, 4, 2, 7, 10, 8, 3, 10],
    [9, 6, 1, 4, 7, 9, 7, 8]
])

wx = np.array([0.25, 0.35, 0.2, 0.1, 0.1])
```

**The upper half of the decision matrix as matrix y and the corresponding weights from the first column as vector wy**

```python
y = np.array([
    [2, 10, 10, 10, 1, 5, 2, 9],
    [3, 10, 4, 6, 2, 4, 3, 8],
    [4, 10, 8, 7, 3, 3, 1, 10],
    [9, 8, 3, 9, 4, 5, 3, 9],
    [1, 8, 6, 2, 1, 4, 4, 8]
])

wy = np.array([0.25, 0.25, 0.25, 0.15, 0.1])
```

**The weighted scores for each segment**

```python
a = 0
x_weighted_scores = []
for j in range(8):
  for i in range(5):
    a = a + (x[i][j]*wx[i])
  x_weighted_scores.append(a)
  a = 0
```

```python
a = 0
y_weighted_scores = []
for j in range(8):
  for i in range(5):
    a = a + (y[i][j]*wy[i])
  y_weighted_scores.append(a)
  a = 0
```

**Segment evaluation plot**

```python
fig, ax = plt.subplots()
size = 8
ax.plot(np.arange(size), x_weighted_scores, 'r', label='Upper Half')
ax.plot(np.arange(size), y_weighted_scores, 'b', label='Lower Half')
ax.legend()
plt.xlabel('Segment')
plt.ylabel('Weighted Score')
plt.title('Segment Evaluation Plot')
plt.show()
```
